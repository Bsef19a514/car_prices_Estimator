# Handling missing values

# cleaning our df

```python
df2['Year'].unique()
```

```python
# Removing the cars that are droven more than 50000 km and there model is older than 2016 and they
#say they are new ...beacause they are not new anymore
df2[(df2['Year']<=2015)& (df2['Condition']=='New')&(df2['KMs Driven']>=50000)]
```

```python
df3=df2[~((df2['Year']<=2015)& (df2['Condition']=='New')&(df2['KMs Driven']>=50000))]
df3
```

```python
#Removing all those models whose count is <=5
df3['Model']=df3['Model'].apply(lambda x: x.strip())
df4=df3.groupby(df3['Model'])
model_count=df4['Model'].agg("count").sort_values(ascending=False)
model_count
```

```python
len(model_count[model_count<=10])
```

```python
model_count_lessthan_6=model_count[model_count<=10]
model_count_lessthan_6
```

```python
df5=df3.copy()
df5
```

```python
df5['Model']=df5['Model'].apply(lambda x: 'other' if x in model_count_lessthan_6 else x)
df5
```

# Converting Model columns to label encoder

```python
model_dummies_df=pd.get_dummies(df7['Model'])
model_dummies_df
```

```python
model_dummies_df=model_dummies_df.drop('other',axis='columns')
model_dummies_df
```

```python
cond_dummies_df=pd.get_dummies(df7['Condition'])
cond_dummies_df
```

```python
fuel_dummies_df=pd.get_dummies(df7['Fuel'])
fuel_dummies_df
```

```python
df8=pd.concat([df7,cond_dummies_df,fuel_dummies_df,model_dummies_df],axis='columns')
df8
```

```python
df9=df8.drop(["Condition","Fuel","Model"],axis='columns')
df9
```

# Choosing best fit model

```python
from sklearn import linear_model
```

```python
model_parameters={
     'Linear_regression':
     {
         "clf":LinearRegression(),'parameters':{'normalize':[True,False]}
     },
     'Linear_model.Lasso':
     {
         "clf":linear_model.Lasso(),"parameters":{'alpha':[1.0],"selection" : ['cyclic', 'random']}
     },
    'Linear_model.Ridge':
     {
         "clf":linear_model.Ridge(),"parameters":{'alpha':[1.0],
                                                  "solver" : ['auto', 'svd' ]}
     },
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
```

```python
cv=ShuffleSplit(n_splits=5,test_size=0.3,random_state=40)
```

```python
scores=[]
for model_name,mp in model_parameters.items():
    clf=GridSearchCV(mp['clf'],mp['parameters'],cv=cv,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'Model':model_name,'Best parametrs':clf.best_params_,'Best_score':clf.best_score_
    })
```

```python
scores_df=pd.DataFrame(scores)
scores_df
```

# Choosing my linear model to be my final model

```python
def predict_price(model,km_driven,year,fuel,condition):
    index_model=np.where(X.columns==model)[0][0]
    index_fuel=np.where(X.columns==fuel)[0][0]
    index_condition=np.where(X.columns==condition)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=km_driven
    x[1]=year
    if index_model>=0:
        x[index_model]=1
    if index_fuel>=0:
        x[index_fuel]=1
    if index_condition>=0:
        x[index_condition]=1
    return lr_model.predict([x])[0]
```
